```python
def fix_wicket_type(series):
    data = []
    if series[0] == 'b':
        data.append('bowled')
    elif series[0] == 'r':
        data.append('run out')
    elif series[0] == 'c':
        data.append('catch')
    elif series[0] == 'l':
        data.append('lbw')
    elif series[0] == 's':
        data.append('stumped')
    elif series[0] == 'h':
        data.append('hit wicket')
    elif series[0] == 'o':
        data.append('obstructing the field')
    else:
        data.append('not out')
    return data[0]
```

```python
df.wicket_type = df.wicket_taken.apply(fix_wicket_type)
```

```python
# Test

df.sample(5)
```

#### derived column from wicket_taken

```python
# add column for wicket taker

df.insert(8,'wicket_taken_by',np.nan)
```

```python
def fix_wicket_taken(s):
    data = []
    
    if 'not out' in s:
        data.append('not out')
    elif 'run out' in s:
        data.append('run out') 
    elif ' b ' in s or 'b ' in s:
        # take text after ' b '
        s = s[s.find(' b ')+2:].strip()
        s = s[1:].strip() if s[0:2] == 'b ' else s
        data.append(s)
    else:
        data.append(s)
    return data[0]
```

```python
df['wicket_taken_by'] = df.wicket_taken.apply(fix_wicket_taken)
```

```python
# Test

df.sample(5)
```

```python
df.drop(columns = 'wicket_taken', inplace=True)
```

```python
# For better Column Name

df.rename(columns={
    'captain': 'match_captain',
    'wicket_taken_by': 'dismissed_by',
    'stadium': 'venue',
    'team': 'team_name',
    'opposing_team': 'opponent_team'
}, inplace=True)
```
