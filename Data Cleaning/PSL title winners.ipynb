```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
DATA_URL = 'https://www.crictracker.com/pakistan-super-league-winners-and-runners-list/'
 
# Prepare a dictionary with the User-Agent header
# and its value (UA string for Firefox)
request_headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/111.0'
}


response = requests.get(DATA_URL, headers=request_headers)
 
# Pass the page HTML (response.text) as input
html_tables = pd.read_html(response.text, index_col=0, header=0)

# first table table winner
title_winner = html_tables[0]

# second table player of series
player_series = html_tables[1]
```

```python
title_winner = title_winner.drop(10, axis=0).reset_index(drop=True).sort_values(by='Year')
```

```python
player_of_series = player_series.drop(2025, axis=0).reset_index().sort_values(by='Year')
player_of_series
```

```python
title_winner.Year=title_winner.Year.astype(int)
```

```python
df = player_of_series.merge(title_winner, how= 'inner', on='Year').drop(columns = 'Winner_y')
df.rename(columns = {'Winner_x':'Winner'}, inplace=True)
```

```python
df
```

```python
df = df[['Year', 'Winner','Won By', 'Runner-up', 'Captain', 'Man of the Match', 'Player of the Tournament','Host']]
```
